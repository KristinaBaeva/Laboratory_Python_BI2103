–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:

1) –ü–æ–¥–∫–ª—é—á–∞–µ–º –º–æ–¥—É–ª—å random, –∏–∑ –∫–æ—Ç–æ—Ä–æ–≥–æ –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é (–º–µ—Ç–æ–¥) randint(). Python –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ª—É—á–∞–π–Ω–æ–µ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è –Ω–∏–∂–Ω–∏–º–∏ –∏ –≤–µ—Ä—Ö–Ω–∏–º–∏ –ø—Ä–µ–¥–µ–ª–∞–º–∏, –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–º–∏, –∫–∞–∫ –¥–≤–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞;

4) –ò—Å–ø–æ–ª—å–∑—É–µ–º –≤–Ω—É—Ç—Ä–∏ —Ç–µ–ª–∞ —Ü–∏–∫–ª–∞ –º–µ—Ç–æ–¥ append() –¥–æ–±–∞–≤–ª—è–µ—Ç –≤ –∫–æ–Ω–µ—Ü —Å–ø–∏—Å–∫–∞ —ç–ª–µ–º–µ–Ω—Ç, –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–π –µ–º—É –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞. –í –¥–∞–Ω–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ –Ω–∞–º –Ω—É–∂–Ω—ã —á–∏—Å–ª–∞ –æ—Ç 0 –¥–æ 100.

5) –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é chunk_list (chunk - –ø–æ—Ä–µ–∑–∞–Ω–Ω—ã–π –Ω–∞ –∫—É—Å–æ—á–∫–∏). –í –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ –º—ã –ø–µ—Ä–µ–¥–∞–µ–º –Ω–∞—à —Å–ø–∏—Å–æ–∫ random_list –∏ –∑–∞—è–≤–ª–µ–Ω–Ω—ã–µ –≤ –∑–∞–¥–∞–Ω–∏–∏ k - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç –¥–µ–ª–∏—Ç—å —Å–ø–∏—Å–æ–∫ random_list –Ω–∞ —á–∞—Å—Ç–∏.

6) –í —Ç–µ–ª–µ —Ñ—É–Ω–∫—Ü–∏–∏ –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è —Ü–∏–∫–ª for, –≤ –∫–æ—Ç–æ—Ä–æ–º –º—ã –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –Ω–∞—à —Å–ø–∏—Å–æ–∫ random_list, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏—é range(). –ú—ã –Ω–∞—á–∏–Ω–∞–µ–º –æ—Ç–±–æ—Ä —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è 0, –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–π –¥–ª–∏–Ω–µ –Ω–∞—à–µ–≥–æ random_list —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ len(), —Ä–∞–∑–±–∏–≤–∞—è –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é - k_elems.

7) –ö–æ–≥–¥–∞ –º—ã —Ç–µ—Å—Ç–∏—Ä—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –æ–±—ä—è–≤–ª—è–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é k_elems, –∫–æ—Ç–æ—Ä–æ–π –º—ã —É—Å—Ç–∞–Ω–æ–≤–∏–ª–∏ –ª—é–±–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —á—Ç–æ–±—ã —É–∫–∞–∑–∞—Ç—å, —á—Ç–æ –º—ã —Ö–æ—Ç–∏–º —Ä–∞–∑–¥–µ–ª–∏—Ç—å –Ω–∞—à —Å–ø–∏—Å–æ–∫ –Ω–∞ —Ñ—Ä–∞–≥–º–µ–Ω—Ç—ã —Ä–∞–∑–º–µ—Ä–æ–º k.

8) –ó–∞—Ç–µ–º –º—ã –∑–∞–Ω–æ—Å–∏–º –≤ –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ chunked_list (–∫–æ—Ç–æ—Ä—ã–π –º—ã –æ–±—ä—è–≤–∏–ª–∏ –≤—ã—à–µ) —Å –ø–æ–º–æ—â—å—é —É–∂–µ –∏–∑–≤–µ—Å—Ç–Ω–æ–π –º–Ω–µ —Ñ—É–Ω–∫—Ü–∏–∏ append() —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞ —Å –ø–æ–º–æ—â—å—é –∑–∞–ø–∏—Å–∏ "i:i + k_elems", –∫–æ—Ç–æ—Ä–∞—è –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –±—É–¥–µ—Ç –≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å—Å—è n-—Å–ø–∏—Å–æ–∫ –æ—Ç i –¥–æ i + —Å–ª–µ–¥—É—é—â–∏–π n-—Å–ø–∏—Å–æ–∫

9) –û–±—ä—è–≤–ª—è–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é result, –≤ –∫–æ—Ç–æ—Ä—É—é –∑–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–±–æ—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏ chunk_list

````python
from random import randint #–∏—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ randint()

######################################################

def chunk_list(random_list, k_elems): #—Å–æ–∑–¥–∞—ë–º —Ñ—É–Ω–∫—Ü–∏—é chunk_list
    chunked_list = []
    for i in range(0, len(random_list), k_elems):
        chunked_list.append(random_list[i:i + k_elems])
    return chunked_list

######################################################

random_list = [] #—Å–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫ 
for i in range(20): #–¥–∞–ª–µ–µ –Ω–∞–º –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è —Ü–∏–∫–ª for, –≤ –∫–æ—Ç–æ—Ä–æ–º –º—ã –±—É–¥–µ–º –∑–∞–ø–æ–ª–Ω—è—Ç—å —Å–ø–∏—Å–æ–∫ —Ä–∞–Ω–¥–æ–º–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
    random_list.append(randint(0, 100))
#print(random_list)

k_elems = 6
result = chunk_list(random_list, k_elems)
print(result)
````

2) –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –≤—ã—á–∏—Å–ª—è—é—â—É—é —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞, –ø–µ—Ä–µ–¥–∞–≤–∞–µ–º–æ–≥–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞.

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:

2) –í—ã–ø–æ–ª–Ω–∏–º –¥–∞–Ω–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ —Å –ø–æ–º–æ—â—å—é —Ä–µ–∫—É—Ä—Å–∏–∏. –†–µ–∫—É—Ä—Å–∏—è - —ç—Ç–æ —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–∑—ã–≤–∞–µ—Ç —Å–∞–º—É —Å–µ–±—è. 

3) –û–±—ä—è–≤–ª—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é factorial –∏ –ø–µ—Ä–µ–¥–∞—ë–º –≤ –Ω–µ—ë –∞—Ä—É–≥–º–µ–Ω—Ç number (—ç—Ç–æ —Ç–æ —Å–∞–º–æ–µ —á–∏—Å–ª–æ, —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª –∫–æ—Ç–æ—Ä–æ–≥–æ –º—ã –ø—ã—Ç–∞–µ–º—Å—è –Ω–∞–π—Ç–∏)

3) –†–µ–∫—É—Ä—Å–∏—è –∑–∞–∫–∞—á–Ω–∏–≤–∞–µ—Ç —Å–≤–æ—é —Ä–∞–±–æ—Ç—É, –∫–∞–∫ —Ç–æ–ª—å–∫–æ number —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Ä–∞–≤–Ω—ã–π –µ–¥–∏–Ω–∏—Ü–µ, –ø–æ—Ç–æ–º—É —á—Ç–æ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª –µ–¥–∏–Ω–∏—Ü—ã - —ç—Ç–æ 1, –º—ã –∏ –ø–∏—à–µ–º –≤ –ø–µ—Ä–≤–æ–º if: –µ—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–≤–Ω–æ 1 - —Ç–æ –º—ã –≤–µ—Ä–Ω—ë–º –∑–Ω–∞—á–µ–Ω–∏–µ number. –§—É–Ω–∫—Ü–∏—è –ø—Ä–µ–∫—Ä–∞—â–∞–µ—Ç—Å—è, –∏ —Ä–µ–∫—É—Ä—Å–∏—è –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç—Å—è.

4) –ï—Å–ª–∏ –∂–µ –Ω–∞—à–µ —á–∏—Å–ª–æ –Ω–µ —Ä–∞–≤–Ω–æ 1, —Ç–æ –º—ã —É–º–Ω–æ–∂–∞–µ–º –∑–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ number –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é factorial (number - 1). (number - 1) - —ç—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ —á–∏—Å–ª–∞, —É–º–µ–Ω—å—à–∞—é—â–µ–µ—Å—è –Ω–∞ 1, –∫–∞–∫ –≤ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–º –≤—ã—à–µ –ø—Ä–∏–º–µ—Ä–µ —Å —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–æ–º –ø—è—Ç–∏. –°–Ω–∞—á–∞–ª–∞ number (5) —É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é factorial (5 - 1). –ò —Ç–∞–∫ —Ñ—É–Ω–∫—Ü–∏—è –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å –ø–æ–∫–∞ —á–∏—Å–ª–æ (number) –≤ —Å–∫–æ–±–∫–∞—Ö –Ω–µ —Å—Ç–∞–Ω–µ—Ç —Ä–∞–≤–Ω–æ 1, –∏ —Ä–µ–∫—É—Ä—Å–∏—è –Ω–µ –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è.

5) –ó–∞–¥–∞–µ–º –ª—é–±–æ–µ –∂–µ–ª–∞–µ–º–æ–µ —á–∏—Å–ª–æ –∏ –≤—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

````python
def factorial(number):
    if number == 1:
        return number
    else:
        return number * factorial(number - 1)

task_number = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ: '))
print(factorial(task_number))
````

3) –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —É–º–Ω–æ–∂–µ–Ω–∏—è, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ —Ç—Ä–µ—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤. –§—É–Ω–∫—Ü–∏—é –≤—ã–∑–≤–∞—Ç—å —Å –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–º–∏ –Ω–∏–∂–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏. –î–ª—è —Å–ª—É—á–∞—è a4 –≤—ã–±—Ä–∞—Ç—å 3 –ø–µ—Ä–≤—ã—Ö –∏ 3 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è.

a1 = (15, 10 ,5)

a2 = (3, 1)

a3 = [2, 35, 55]

a4 = (5, 10 ,15, 20)

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:
1) –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ Func —Å –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏ *args* —á—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤.
2) –ü—Ä–æ–≤–µ—Ä–∫–∞, –µ—Å–ª–∏ –¥–ª–∏–Ω–∞ args —Ä–∞–≤–Ω–∞ 0 (—Ç.–µ., –µ—Å–ª–∏ –Ω–∏ –æ–¥–Ω–æ–≥–æ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ –Ω–µ –ø–µ—Ä–µ–¥–∞–Ω–æ), —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç 0.
3) –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é key_arg (–∫–ª—é—á–µ–≤–æ–π –∞—Ä–≥—É–º–µ–Ω—Ç, –∫–æ—Ç–æ—Ä—ã–π –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è —É–º–Ω–æ–∂–µ–Ω–∏—è) —Ä–∞–≤–Ω–æ–π 1.
4) –ó–∞—Ç–µ–º –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Ü–∏–∫–ª for, –∫–æ—Ç–æ—Ä—ã–π –ø–µ—Ä–µ–±–∏—Ä–∞–µ—Ç –≤—Å–µ –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–µ –∞—Ä–≥—É–º–µ–Ω—Ç—ã.
5) –í –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –∞—Ä–≥—É–º–µ–Ω—Ç (arg) —É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ key_arg, —á—Ç–æ–±—ã –Ω–∞–∫–æ–ø–∏—Ç—å –∏—Ö –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ.
5) –í –∫–æ–Ω—Ü–µ —Ü–∏–∫–ª–∞ key_arg —Å–æ–¥–µ—Ä–∂–∏—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –≤—Å–µ—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤.
6) –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ key_arg.

*Func(5) –≤–µ—Ä–Ω–µ—Ç 5 (–ø–æ—Å–∫–æ–ª—å–∫—É –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç —É–º–Ω–æ–∂–∞–µ—Ç—Å—è —Å–∞–º –Ω–∞ —Å–µ–±—è).

````python
def Func(*args):

    if len(args) == 0:  
        return(0)       
                    
    key_arg = 1
    for arg in args:
        key_arg *= arg
    return(key_arg)


print(Func("–í —Å–ª—É—á–∞–µ –µ—Å–ª–∏ –º—ã –Ω–∏—á–µ–≥–æ –Ω–µ –ø–µ—Ä–µ–¥–∞—ë–º: "))
print("a1:", Func(15, 10, 5))
print("a2:", Func(3, 1))
print("a3:", Func(2, 35, 55))
print("a4:", Func(5, 10, 15, 10, 15, 20))
print("a5:", Func(5))
````

4) –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞ r –≤—ã—á–∏—Å–ª—è–µ—Ç –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ—Ä—Ç–µ–∂ –∏–∑ 2 –∑–Ω–∞—á–µ–Ω–∏–π.

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:
1) –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è Pi –∫–∞–∫ 3.14.
2) –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ radius_characteristics —Å –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º Rad.
3) –í–Ω—É—Ç—Ä–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∫—Ä—É–≥–∞ S –∏ –¥–ª–∏–Ω—ã –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ C –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞ Rad:
4) –ü–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ S —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ Pi * Rad**2.
5) –î–ª–∏–Ω–∞ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ C —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ 2 * Pi * Rad.
6) –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ [S, round(C, 1)], –≥–¥–µ S - –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞, –∞ C - –¥–ª–∏–Ω–∞ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏, –æ–∫—Ä—É–≥–ª–µ–Ω–Ω–∞—è –¥–æ –æ–¥–Ω–æ–π –¥–µ—Å—è—Ç–æ–π.


````python
Pi = 3.14

def radius_characteristics (Rad):
    S = Pi * Rad**2
    C = 2*Pi*Rad
    return [S, round(C, 1)]
    
result = radius_characteristics(5)
print (result)    
````

5) –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π, –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—â–∏—Ö —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –≤ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. 
–°–æ–∑–¥–∞—Ç—å —Å–ª–æ–≤–∞—Ä—å —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∏–º —Å–∏–º–≤–æ–ª–∞–º–∏ –æ–ø–µ—Ä–∞—Ü–∏–π. 
–î–ª—è –¥–≤—É—Ö –∑–∞—Ä–∞–Ω–µ–µ –∑–∞–¥–∞–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, 25 –∏ 4) –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤—ã–±—Ä–∞–Ω–Ω—É—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –æ–ø–µ—Ä–∞—Ü–∏—é. 

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:
1) –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–µ—Ç—ã—Ä–µ—Ö —Ñ—É–Ω–∫—Ü–∏–π: Addition, Subtraction, Multiplication –∏ Division, –∫–∞–∂–¥–∞—è –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ a –∏ b –∏ –≤—ã–ø–æ–ª–Ω—è–µ—Ç –æ–ø–µ—Ä–∞—Ü–∏–∏ —Å–ª–æ–∂–µ–Ω–∏—è, –≤—ã—á–∏—Ç–∞–Ω–∏—è, —É–º–Ω–æ–∂–µ–Ω–∏—è –∏ –¥–µ–ª–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ö–∞–∂–¥–∞—è –∏–∑ —ç—Ç–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ–ø–µ—Ä–∞—Ü–∏–∏, –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã–π –≤ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (—Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º int).

2) –ó–∞–ø—Ä–æ—Å –≤–≤–æ–¥–∞ –¥–≤—É—Ö —á–∏—Å–µ–ª a –∏ b —Å –ø–æ–º–æ—â—å—é input(). –û–±–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—Ç—Å—è –≤ —á–∏—Å–ª–∞ —Å –ø–ª–∞–≤–∞—é—â–µ–π –∑–∞–ø—è—Ç–æ–π (float).

3) –ó–∞–ø—Ä–æ—Å –≤–≤–æ–¥–∞ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ (operator) —Å –ø–æ–º–æ—â—å—é input().

4) –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è dictionary, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–æ–º—É –æ–ø–µ—Ä–∞—Ç–æ—Ä—É ('+', '-', '*', '/') —Å–æ–ø–æ—Å—Ç–∞–≤–ª—è–µ—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è –∏–∑ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —Ä–∞–Ω–µ–µ (Addition, Subtraction, Multiplication, Division).

5) –í—ã–∑–æ–≤ –≤—ã–±—Ä–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ —Å–ª–æ–≤–∞—Ä—è dictionary[operator](a, b) —Å –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏ a –∏ b. –í –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≤–≤–µ–¥–µ–Ω–Ω–æ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞, –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω–µ–Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è —Å —á–∏—Å–ª–∞–º–∏ a –∏ b.


````python
def Addition(a, b):
    return int(a + b)

def Subtraction (a, b):
    return int(a - b)

def Multiplication (a, b):
    return int(a * b)

def Division (a, b):
    return int(a // b)

a, b, operator = float(input()), float(input()), input() 

dictionary = {'+': Addition,
              '-': Subtraction,
              '*': Multiplication,
              '/': Division}

print(dictionary[operator](a, b))
````

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:

2) –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏ –Ω–∞ —ç–ª–µ–º–µ–Ω—Ç—ã. –°—Ç—Ä–æ–∫–∞ expression —Ä–∞–∑–¥–µ–ª—è–µ—Ç—Å—è –Ω–∞ —ç–ª–µ–º–µ–Ω—Ç—ã (—á–∏—Å–ª–æ1, –æ–ø–µ—Ä–∞—Ç–æ—Ä, —á–∏—Å–ª–æ2) —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ split().

3) –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ñ–æ—Ä–º–∞—Ç–∞ –≤–≤–æ–¥–∞. –ü—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Ä–∞–≤–Ω–æ 3 (—á–∏—Å–ª–æ1, –æ–ø–µ—Ä–∞—Ç–æ—Ä, —á–∏—Å–ª–æ2). –ï—Å–ª–∏ –Ω–µ —Ä–∞–≤–Ω–æ, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è —Å–æ–æ–±—â–µ–Ω–∏–µ "–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞".

4) –ü—Ä–æ–≤–µ—Ä–∫–∞ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –ü—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è, —á—Ç–æ —á–∏—Å–ª–æ1 –∏ —á–∏—Å–ª–æ2 —Å–æ—Å—Ç–æ—è—Ç –∏–∑ —Ü–∏—Ñ—Ä (—è–≤–ª—è—é—Ç—Å—è —á–∏—Å–ª–∞–º–∏) —Å  –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ isdigit(). –ï—Å–ª–∏ –Ω–µ —è–≤–ª—è—é—Ç—Å—è, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è —Å–æ–æ–±—â–µ–Ω–∏–µ "–í–≤–µ–¥–∏—Ç–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞". 

5) –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫ –≤ —á–∏—Å–ª–∞. –°—Ç—Ä–æ–∫–æ–≤—ã–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è —á–∏—Å–ª–∞1 –∏ —á–∏—Å–ª–æ2 –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—Ç—Å—è –≤ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

6) –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞:

7) –í –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ (+, -, *, /), –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è –Ω–∞–¥ —á–∏—Å–ª–∞–º–∏.

–î–ª—è –¥–µ–ª–µ–Ω–∏—è (/) –ø—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è, —á—Ç–æ –¥–µ–ª–∏—Ç–µ–ª—å (—á–∏—Å–ª–æ2) –Ω–µ —Ä–∞–≤–µ–Ω –Ω—É–ª—é.
–î–ª—è —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è (//), –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ.

**–ú–µ—Ç–æ–¥ split () —Ä–∞–∑–±–∏–≤–∞–µ—Ç –æ–±—ä–µ–∫—Ç String –Ω–∞ –º–∞—Å—Å–∏–≤ —Å—Ç—Ä–æ–∫ –ø—É—Ç—ë–º —Ä–∞–∑–¥–µ–ª–µ–Ω–∏—è —Å—Ç—Ä–æ–∫–∏ —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–æ–¥—Å—Ç—Ä–æ–∫–æ–π.

````python
def WordCalculator(expression):
    
    elements = expression.split()
    
    if len(elements) != 3:
        return "–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞"
    
    num1, op, num2 = elements
    
    if not (num1.isdigit() and num2.isdigit()):
        return "–í–≤–µ–¥–∏—Ç–µ —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞"
    
    num1, num2 = int(num1), int(num2)

    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    elif op == '*':
        result = num1 * num2
    elif op == '/':
        if num2 == 0:
            return "–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ"
        result = num1 // num2
    else:
        return "–ù–µ–ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è"
    
    return result

expression = input("–í–≤–µ–¥–∏—Ç–µ —Å—Ç—Ä–æ–∫–æ–≤–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≤–∏–¥–∞ '<—á–∏—Å–ª–æ> <–æ–ø–µ—Ä–∞—Ü–∏—è> <—á–∏—Å–ª–æ>': ")
print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", WordCalculator(expression))
````

7) –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–æ—Ç 0 –¥–æ 999) –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –≤ —á–∏—Å–ª–æ —Ç–∏–ø–∞ int. –ü—Ä–∏–º–µ—Ä: to_int('—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏') -> 33

**1) –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è —Å —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–º–∏ –∏ —á–∏—Å–ª–∞–º–∏.

–°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å dictionary, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–∞–ø–∏—Å–∞–Ω–Ω—ã–µ —Å–ª–æ–≤–∞–º–∏ —á–∏—Å–ª–∞.

**2) –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ —Ç–µ–∫—Å—Ç–∞ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏:

–í—Ö–æ–¥–Ω–æ–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (–ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è) text_number —Ä–∞–∑–±–∏–≤–∞–µ—Ç—Å—è –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ .split(). –°–ø–ª–∏—Ç —Ä–∞–±–æ—Ç–∞–µ—Ç –ø–æ –ø—Ä–∏–Ω—Ü–∏–ø—É: –∫–æ–≥–¥–∞ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –ø–µ—Ä–≤—ã–π –ø—Ä–æ–±–µ–ª, –º–µ—Ç–æ–¥ —Ä–∞–∑–±–∏–≤–∞–µ—Ç —Å—Ç—Ä–æ–∫—É.
–í—Å–µ —Å–ª–æ–≤–∞ –ø—Ä–∏–≤–æ–¥—è—Ç—Å—è –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É —Å –ø–æ–º–æ—â—å—é .lower(), —á—Ç–æ–±—ã –æ–±–µ—Å–ø–µ—á–∏—Ç—å –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –æ—Ç —Ä–µ–≥–∏—Å—Ç—Ä–∞ –ø—Ä–∏ —Å–æ–ø–æ—Å—Ç–∞–≤–ª–µ–Ω–∏–∏ —Å–ª–æ–≤ —Å —á–∏—Å–ª–∞–º–∏.

**3) –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö:

–°–æ–∑–¥–∞—ë–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é current_number, –∫–æ—Ç–æ—Ä—É—é –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º 0.
current_number –±—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Ç–µ–∫—É—â–µ–π —Å—É–º–º—ã —á–∏—Å–µ–ª –ø–æ –º–µ—Ä–µ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –ø–æ —Å–ª–æ–≤–∞–º.

**4) –ò—Ç–µ—Ä–∞—Ü–∏—è –ø–æ —Å–ª–æ–≤–∞–º –∏ —Å—É–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π:

–°–ª–µ–¥—É—é—â–∏–º —à–∞–≥–æ–º –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–µ—Ä–µ–±–æ—Ä –ø–æ –∫–∞–∂–¥–æ–º—É —Å–ª–æ–≤—É, –∫–æ—Ç–æ—Ä–æ–µ –±—ã–ª–æ —Ä–∞–∑–±–∏—Ç–æ –Ω–∞ —ç—Ç–∞–ø–µ 2.
–î–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–ª–æ–≤–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –ø—Ä–æ–≤–µ—Ä–∫–∞: –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –ª–∏ –æ–Ω–æ –≤ —Å–ª–æ–≤–∞—Ä–µ dictionary. –ï—Å–ª–∏ —Å–ª–æ–≤–æ –Ω–∞–π–¥–µ–Ω–æ –≤ —Å–ª–æ–≤–∞—Ä–µ, –µ–≥–æ —á–∏—Å–ª–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–∏–±–∞–≤–ª—è–µ—Ç—Å—è –∫ current_number.
–ï—Å–ª–∏ —Å–ª–æ–≤–æ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ –≤ —Å–ª–æ–≤–∞—Ä–µ (—Å–ª–æ–≤–æ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–º), —Ñ—É–Ω–∫—Ü–∏—è —Å—Ä–∞–∑—É –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É "–ù–µ–≤–µ—Ä–Ω–æ–µ —Å–ª–æ–≤–æ –≤ —á–∏—Å–ª–µ". –≠—Ç–æ –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–∞–µ—Ç –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏, –µ—Å–ª–∏ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ –Ω–µ–≤–µ—Ä–Ω–æ–µ —Å–ª–æ–≤–æ.

**5)–í–æ–∑–≤—Ä–∞—â–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:

–ü–æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏–∏ –∏—Ç–µ—Ä–∞—Ü–∏–∏ –ø–æ —Å–ª–æ–≤–∞–º –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ current_number, –∫–æ—Ç–æ—Ä–æ–µ —è–≤–ª—è–µ—Ç—Å—è —Å—É–º–º–æ–π —á–∏—Å–µ–ª, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ç–µ–∫—Å—Ç–æ–≤–æ–º—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—é.

````python
def to_int(text_number):
    # 1
    dictionary = {
        '–Ω–æ–ª—å': 0, '–æ–¥–∏–Ω': 1, '–¥–≤–∞': 2, '—Ç—Ä–∏': 3, '—á–µ—Ç—ã—Ä–µ': 4,
        '–ø—è—Ç—å': 5, '—à–µ—Å—Ç—å': 6, '—Å–µ–º—å': 7, '–≤–æ—Å–µ–º—å': 8, '–¥–µ–≤—è—Ç—å': 9,
        '–¥–µ—Å—è—Ç—å': 10, '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å': 11, '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å': 12, '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å': 13, '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å': 14,
        '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 15, '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å': 16, '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 17, '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 18, '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 19,
        '–¥–≤–∞–¥—Ü–∞—Ç—å': 20, '—Ç—Ä–∏–¥—Ü–∞—Ç—å': 30, '—Å–æ—Ä–æ–∫': 40, '–ø—è—Ç—å–¥–µ—Å—è—Ç': 50, '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç': 60,
        '—Å–µ–º—å–¥–µ—Å—è—Ç': 70, '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç': 80, '–¥–µ–≤—è–Ω–æ—Å—Ç–æ': 90,
        '—Å—Ç–æ': 100, '–¥–≤–µ—Å—Ç–∏': 200, '—Ç—Ä–∏—Å—Ç–∞': 300, '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞': 400, '–ø—è—Ç—å—Å–æ—Ç': 500,
        '—à–µ—Å—Ç—å—Å–æ—Ç': 600, '—Å–µ–º—å—Å–æ—Ç': 700, '–≤–æ—Å–µ–º—å—Å–æ—Ç': 800, '–¥–µ–≤—è—Ç—å—Å–æ—Ç': 900
    }

    # 2
    words = text_number.lower().split()

    #3
    current_number = 0

    # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å–ª–æ–≤–∞–º –∏ —Å—É–º–º–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è
    for word in words:
        if word in dictionary:
            current_number += dictionary[word]
        else:
            # –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–ª—É—á–∞—è, –∫–æ–≥–¥–∞ —Å–ª–æ–≤–æ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ –≤ —Å–ª–æ–≤–∞—Ä–µ
            return "–ù–µ–≤–µ—Ä–Ω–æ–µ —Å–ª–æ–≤–æ –≤ —á–∏—Å–ª–µ"

    return current_number

#text_number = '–ø—è—Ç—å—Å–æ—Ç –¥–≤–∞–¥—Ü–∞—Ç—å —Ç—Ä–∏'
text_number = input()
result = to_int(text_number)

print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", result)
````

9) –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ (–æ—Ç 0 –¥–æ 999) –∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –≤ —á–∏—Å–ª–æ —Ç–∏–ø–∞ int. –ò —Å–æ–æ–±—â–∞–µ—Ç –æ–± –æ—à–∏–±–∫–∞—Ö (–≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –æ–ø–∏—Å–∞–Ω–∏–µ —Ç–∏–ø–∞ –æ—à–∏–±–∫–∏ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —á–∏—Å–ª–æ -1). –ü—Ä–∏–º–µ—Ä: to_int('—Ç—Ä–∏–¥—Ü–∞—Ç—å —Ç—Ä–∏') -> 33 –ü—Ä–∏–º–µ—Ä: to_int("—Å—Ç–æ —Å–æ—Ä–æ–∫ —Ç—Ä–∏–¥—Ü–∞—Ç—å –¥–≤–∞ —Å—Ç–æ") -> -1 –í—ã–≤–æ–¥: —Ç—Ä–∏–¥—Ü–∞—Ç—å - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ —á–∏—Å–ª–µ.

–ê–ª–≥–æ—Ä–∏—Ç–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è:
1) –°–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å numbers_dict, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–∞–ø–∏—Å–∞–Ω–Ω—ã–µ —Å–ª–æ–≤–∞–º–∏ —á–∏—Å–ª–∞. –ù–∞–ø—Ä–∏–º–µ—Ä, 1 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç '–æ–¥–∏–Ω'.

2) –§—É–Ω–∫—Ü–∏—è int_to_text(number) –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç - number, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å.

3) –ï—Å–ª–∏ –ø–µ—Ä–µ–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ number —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –≤ —Å–ª–æ–≤–∞—Ä–µ numbers_dict, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –¥–ª—è —ç—Ç–æ–≥–æ —á–∏—Å–ª–∞. –ü—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é —É—Å–ª–æ–≤–Ω–æ–π –æ–ø–µ—Ä–∞—Ü–∏–∏ if.

4) –ï—Å–ª–∏ —á–∏—Å–ª–æ number –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç 1 –¥–æ 999 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ), —Ç–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–µ–µ:

    hundreds –≤—ã—á–∏—Å–ª—è–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Ç–µ–Ω –≤ —á–∏—Å–ª–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —á–∏—Å–ª–∞ 523 —ç—Ç–æ –±—É–¥–µ—Ç 500).
    
    remainder –≤—ã—á–∏—Å–ª—è–µ—Ç –æ—Å—Ç–∞—Ç–æ–∫, –æ—Å—Ç–∞–≤—à–∏–π—Å—è –ø–æ—Å–ª–µ –≤—ã—á–µ—Ç–∞ —Å–æ—Ç–µ–Ω (–¥–ª—è —á–∏—Å–ª–∞ 523 —ç—Ç–æ –±—É–¥–µ—Ç 23).
    
    –ï—Å–ª–∏ remainder —Ä–∞–≤–µ–Ω 0, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –¥–ª—è —Å–æ—Ç–µ–Ω (–Ω–∞–ø—Ä–∏–º–µ—Ä, '–ø—è—Ç—å—Å–æ—Ç').

    –ï—Å–ª–∏ remainder –º–µ–Ω—å—à–µ 20, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ–º–±–∏–Ω–∞—Ü–∏—é —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–π –¥–ª—è —Å–æ—Ç–µ–Ω –∏ –æ—Å—Ç–∞—Ç–∫–∞ (–Ω–∞–ø—Ä–∏–º–µ—Ä, '–ø—è—Ç—å—Å–æ—Ç –¥–≤–∞–¥—Ü–∞—Ç—å —Ç—Ä–∏').

    –ï—Å–ª–∏ remainder –±–æ–ª—å—à–µ 20, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –¥–∞–ª–µ–µ —Ä–∞–∑–±–∏–≤–∞–µ—Ç remainder –Ω–∞ –¥–µ—Å—è—Ç–∫–∏ –∏ –µ–¥–∏–Ω–∏—Ü—ã:
        tens - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Å—è—Ç–∫–æ–≤ –≤ –æ—Å—Ç–∞—Ç–∫–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è 23 —ç—Ç–æ –±—É–¥–µ—Ç 20).
        ones - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –µ–¥–∏–Ω–∏—Ü –≤ –æ—Å—Ç–∞—Ç–∫–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è 23 —ç—Ç–æ –±—É–¥–µ—Ç 3).
    –ï—Å–ª–∏ ones —Ä–∞–≤–Ω–æ 0, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –¥–ª—è —Å–æ—Ç–µ–Ω –∏ –¥–µ—Å—è—Ç–∫–æ–≤ (–Ω–∞–ø—Ä–∏–º–µ—Ä, '–ø—è—Ç—å—Å–æ—Ç –¥–≤–∞–¥—Ü–∞—Ç—å').

–ï—Å–ª–∏ —á–∏—Å–ª–æ number –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –æ—Ç 1 –¥–æ 999, —Ñ—É–Ω–∫—Ü–∏—è –≤–µ—Ä–Ω–µ—Ç —Å–æ–æ–±—â–µ–Ω–∏–µ "–ß–∏—Å–ª–æ –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ (1-999)".

–ü–æ—Å–ª–µ –≤—ã–∑–æ–≤–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç —á–∏—Å–ª–æ (number) —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã, –∏ —Ñ—É–Ω–∫—Ü–∏—è int_to_text –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è —Å —ç—Ç–∏–º —á–∏—Å–ª–æ–º. –†–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π result.

````python
def int_to_text(number):

    numbers_dict = {
        0: '–Ω–æ–ª—å', 1: '–æ–¥–∏–Ω', 2: '–¥–≤–∞', 3: '—Ç—Ä–∏', 4: '—á–µ—Ç—ã—Ä–µ',
        5: '–ø—è—Ç—å', 6: '—à–µ—Å—Ç—å', 7: '—Å–µ–º—å', 8: '–≤–æ—Å–µ–º—å', 9: '–¥–µ–≤—è—Ç—å',
        10: '–¥–µ—Å—è—Ç—å', 11: '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å', 12: '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å', 13: '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å', 14: '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å',
        15: '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å', 16: '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å', 17: '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å', 18: '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å', 19: '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å',
        20: '–¥–≤–∞–¥—Ü–∞—Ç—å', 30: '—Ç—Ä–∏–¥—Ü–∞—Ç—å', 40: '—Å–æ—Ä–æ–∫', 50: '–ø—è—Ç—å–¥–µ—Å—è—Ç', 60: '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç',
        70: '—Å–µ–º—å–¥–µ—Å—è—Ç', 80: '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç', 90: '–¥–µ–≤—è–Ω–æ—Å—Ç–æ',
        100: '—Å—Ç–æ', 200: '–¥–≤–µ—Å—Ç–∏', 300: '—Ç—Ä–∏—Å—Ç–∞', 400: '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞', 500: '–ø—è—Ç—å—Å–æ—Ç',
        600: '—à–µ—Å—Ç—å—Å–æ—Ç', 700: '—Å–µ–º—å—Å–æ—Ç', 800: '–≤–æ—Å–µ–º—å—Å–æ—Ç', 900: '–¥–µ–≤—è—Ç—å—Å–æ—Ç'
    }

    # number - –ø–µ—Ä–µ–¥–∞–≤–∞–µ–º—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç –≤ –Ω–∞—à—É —Ñ—É–Ω–∫—Ü–∏—é
    if number in numbers_dict:
        return numbers_dict[number]
    
    elif 0 < number < 1000:
        hundreds = number // 100 * 100 #—Ç–æ –µ—Å—Ç—å –µ—Å–ª–∏ –º—ã –≤–≤–µ–ª–∏ 623, 623 // 100, —Ç–æ –ø–æ–ª—É—á–∏–º 6 –∏ —É–º–Ω–æ–∂–∏–≤ –Ω–∞ 100, –ø–æ–ª—É—á–∏–º —Å–æ—Ç–Ω–∏
        remainder = number % 100 #–∑–¥–µ—Å—å –º—ã –Ω–∞—Ö–æ–¥–∏–º —Ç–æ, —á—Ç–æ –æ—Å—Ç–∞–ª–æ—Å—å –æ—Ç —Å–æ—Ç–Ω–∏, —Ç–æ –µ—Å—Ç—å 623%100 = 23
        
        if remainder == 0:
            return numbers_dict[hundreds] #–ï—Å–ª–∏ –æ—Å—Ç–∞—Ç–æ–∫ —Ä–∞–≤–µ–Ω 0, —Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç '—à–µ—Å—Ç—å—Å–æ—Ç').
        
        elif remainder < 20:
            return f"{numbers_dict[hundreds]} {numbers_dict[remainder]}"
        
        else:
            tens = remainder // 10 * 10 #–ø–æ —Ç–æ–º—É –∂–µ –ø—Ä–∏–Ω—Ü–∏–ø—É —Å—á–∏—Ç–∞–µ–º –¥–µ—Å—è—Ç–∫–∏ –∏–∑ –æ—Å—Ç–∞—Ç–∫–∞ remainder 23//10 = 2 * 10 = 20!
            ones = remainder % 10 #–ø–æ–ª—É—á–∏–º 3, —Ç–æ –µ—Å—Ç—å –ø–æ—Å–ª–µ–¥–Ω—é—é —Ü–∏—Ñ—Ä—É
            
            if ones == 0:
                return f"{numbers_dict[hundreds]} {numbers_dict[tens]}"
            else:
                return f"{numbers_dict[hundreds]} {numbers_dict[tens]} {numbers_dict[ones]}"
    else:
        return "–ß–∏—Å–ª–æ –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ (1-999)"

#number = 523
number = int(input())
result = int_to_text(number)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç:", result)
````
